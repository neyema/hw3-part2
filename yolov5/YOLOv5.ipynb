```python
results = pd.read_csv("/content/yolov5/runs/train-cls/exp/results.csv")
results.head()
```

```python
results.columns
```

```python
pyplot.plot(results[' metrics/train_accuracy'], label='train accuracy')
pyplot.plot(results['   metrics/val_accuracy'], label='val accuracy')
pyplot.plot(results['  metrics/test_accuracy'], label='test accuracy')
pyplot.ylabel('accuracy')
pyplot.xlabel('epoch')
pyplot.legend(loc="upper right")
pyplot.show()
```

```python
pyplot.plot(results['     metrics/train_loss'], label='train loss')
pyplot.plot(results['       metrics/val_loss'], label='val loss')
pyplot.plot(results['      metrics/test_loss'], label='test loss')
pyplot.ylabel('loss')
pyplot.xlabel('epoch')
pyplot.legend(loc="upper right")
pyplot.show()
```
